# Meningkatkan akurasi dengan Rolling Averages

### Membagi data set menjadi beberapa data set tergantung team

```python
group_pertandingan= pertandingan.groupby("team")
```

```python
print("Melihat Permainan Manchester City")
group=group_pertandingan.get_group("Manchester City")
group
```

Dalam penggunaan metode Rolling Averages data akan menggunakan hasil dari pertandingan sebelumnya apakah tim tersebut menang atau kalah dan apa yang terjadi dalam pertandingan sebelumnya(3 Pertandingan) yang nantinya akan diolah ke dalam algoritma

```python
def rolling_averages(group, cols, new_col):
    group=group.sort_values("date")
    rolling_stats=group[cols].rolling(3, closed='left').mean()
    group[new_cols]= rolling_stats
    group=group.dropna(subset=new_cols)
    return group
```

```python
cols=["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols=[f"{c}_rolling" for c in cols]
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
pertandingan_rolling = pertandingan.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
pertandingan_rolling
```

```python
pertandingan_rolling=pertandingan_rolling.droplevel("team")
pertandingan_rolling
```

Menghapus Kolom tim agar lebih mudah dalam menghitung algoritma(Kalau tidak harus pakai 2 value)

```python
pertandingan_rolling.index=range(pertandingan_rolling.shape[0])
pertandingan_rolling
```

# Melatih ulang Kode

```python
def buat_prediksi(data, acuan):
    train=data[data["date"] < '2022-01-01']
    test=data[data["date"] > '2022-01-01']
    rf.fit(train[acuan], train["target"])
    prediksi = rf.predict(test[acuan])
    combined=pd.DataFrame(dict(actual=test["target"], predicted=prediksi), index=test.index)
    precision = precision_score(test["target"], prediksi)
    return combined, precision
```

Dalam data ini terdapat 2 tim dimana ada yang menjadi team dan ada yang menjadi opponent untuk itu kita akan melakukan merge agar data lebih akurat.

Akan tetapi dalam data tersebut nama team dan oppenent terkadang tidak sama untuk itu perlu dilakukan nomralize

```python
class MissingDict(dict):
    __missing__ = lambda self, key : key

map_values = {"Brighton and Hove Albion": "Brighton", 
              "Manchester United": "Manchester Utd", 
              "Newcastle United": "Newcastle Utd", 
              "Tottenham Hotspur": "Tottenham", 
              "West Ham United": "West Ham", 
              "Wolverhampton Wanderers": "Wolves"
}
mapping=MissingDict(**map_values)    
```

```python
combined["tim_baru"]= combined["team"].map(mapping)
```

```python
combined
```

```python
merged=combined.merge(combined, left_on=["date", "tim_baru"], right_on=["date", "opponent"])
merged
```
